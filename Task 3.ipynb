```python
import pandas as pd

users = pd.read_csv('/content/users.csv')
photos = pd.read_csv('/content/photos.csv')
likes = pd.read_csv('/content/likes.csv')
comments = pd.read_csv('/content/comments.csv')

print("USERS")
print(users.head())

print("\nPHOTOS")
print(photos.head())

print("\nLIKES")
print(likes.head())

print("\nCOMMENTS")
print(comments.head())
```

```python
print(users.info())

print(photos.info())

print(likes.info())

print(comments.info())
```

```python
print(users.isnull().sum())

print(photos.isnull().sum())

print(likes.isnull().sum())

print(comments.isnull().sum())
```

```python
import matplotlib.pyplot as plt

photos['photo type'].value_counts().plot(kind='bar')
plt.title('Photo Type Distribution')
plt.show()
```

```python
photos['Insta filter used'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Instagram Filter Usage')
plt.show()
```

```python
users['Verified status'].value_counts().plot(
    kind='bar'
)

plt.title('Verified vs Non-Verified Users')
plt.show()
```

```python
likes['like type'].value_counts().head(10).plot(
    kind='bar'
)

plt.title('Top Like Types')
plt.show()
```

```python
comments['Hashtags used count'].value_counts().head(10).plot(
    kind='bar'
)

plt.title('Hashtag Usage Count')
plt.show()
```

```python
print("Total Users:", len(users))
print("Total Photos:", len(photos))
print("Total Likes:", len(likes))
print("Total Comments:", len(comments))
```
